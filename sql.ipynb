```python
import json
import re
import pandas as pd
from urllib.request import Request
from bs4 import BeautifulSoup
import matplotlib.pyplot as plot
from urllib.request import Request, urlopen
```

```python
pages = []

for page_number in range(1, 88):
    with open(f"/Users/mikolajszymanski/Documents/Wizualizacja/Projekt/pracuj/SQL/pracujSQL{page_number}.html", encoding="utf8") as game:
        soup = BeautifulSoup(game, 'html.parser')
        pages.append(soup)
```

```python
print(pages[0].prettify())
```

```python
allOffers = []

for page in pages:
    allOffers = allOffers + page.find_all(class_='ContentBoxstyles__Wrapper-sc-11jmnka-0 ecNPxU JobOfferstyles__ContentBoxWrapper-sc-1rq6ue2-0 lbjnjD OffersListingstyles__OfferWrapper-sc-17xxjbz-0 FvVXP')
```

```python
print(allOffers[0].prettify())
print(len(allOffers))
```

```python
ofertyInfo = []
for i in range(1, 249):
    oferta = {}
    if allOffers[i].find('span', {'data-test': "offer-location"}, class_='JobOfferstyles__DetailsText-sc-1rq6ue2-16 gmwNET') != None:
        miasto = allOffers[i].find('span', {'data-test': "offer-location"}, class_='JobOfferstyles__DetailsText-sc-1rq6ue2-16 gmwNET').text.strip()
    else:
        miasto = 'Nie podano'

    if allOffers[i].find('span', {'data-test': "offer-remote-work"}, class_='JobOfferstyles__DetailsText-sc-1rq6ue2-16 gmwNET') != None:
        if allOffers[i].find('span', {'data-test': "offer-remote-work"}, class_='JobOfferstyles__DetailsText-sc-1rq6ue2-16 gmwNET').text.strip() == 'Praca zdalna':
            zdalnie = 'Tak'
    else:
        zdalnie = 'Nie'

    if allOffers[i].find('span', {'data-test': "offer-salary"}, class_='JobOfferstyles__DetailsText-sc-1rq6ue2-16 eyMfCO') != None:
        xd = str(allOffers[i].find('span', {'data-test': "offer-salary"}, class_='JobOfferstyles__DetailsText-sc-1rq6ue2-16 eyMfCO').text.strip()[:16])
        xd2 = xd.replace('zł', '')
        xd3 = xd2.replace(' ', '')
        xd4 = xd3.replace('/', '')
        xd5 = xd4.replace('brutto', '')
        zarobki = xd5.split('-')
        if len(zarobki) == 2:
            oferta['zarobki_min'] = zarobki[0]
            oferta['zarobki_max'] = zarobki[1]
        else:
            oferta['zarobki_min'] = zarobki[0]
            oferta['zarobki_max'] = zarobki[0]
    else:
        oferta['zarobki_min'] = 'Brak danych'
        oferta['zarobki_max'] = 'Brak danych'
    
    oferta['nazwa'] = allOffers[i].find(class_='JobOfferstyles__TitleHeader-sc-1rq6ue2-4 iWmAcG').text.strip()
    oferta['miasto'] = miasto
    oferta['zdalnie'] = zdalnie

    ofertyInfo.append(oferta)
```
