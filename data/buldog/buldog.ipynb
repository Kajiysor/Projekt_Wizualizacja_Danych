```python
import json
import re
import pandas as pd
import urllib
from urllib.request import Request
from bs4 import BeautifulSoup
import matplotlib.pyplot as plot
from bs4 import BeautifulSoup
from urllib.request import Request, urlopen
```

```python
url = "https://bulldogjob.pl/companies/jobs/s/page,"

pages = []
for page_number in range(1,85):
  print(f'Pobieranie strony {page_number}')
  with urllib.request.urlopen(f'{url}{page_number}') as resp:
    processed_page = BeautifulSoup(resp.read().decode('utf-8'), "html.parser")
    pages.append(processed_page)
```

```python
print(pages[0].prettify())
```

```python
allOffers = []
for page in pages:
  allOffers = allOffers + page.find_all(class_='py-6 md:py-8 px-8 flex flex-wrap relative bg-white mb-2 rounded-lg shadow ')
  allOffers = allOffers + page.find_all(class_='py-6 md:py-8 px-8 flex flex-wrap relative bg-white mb-2 rounded-lg shadow JobListItem_highlight__z1SYx')
```

```python
print(allOffers[2].prettify())

print(len(allOffers))
```

```python
ofertyInfo = []
# print(allOffers[5].find(class_=' group flex rounded-md items-center w-full px-2 my-1 font-normal'))
for i in range(0, 1678):
    oferta = {}
    # print(allOffers[i].find('p', class_='my-3 flex items-center').)
    if allOffers[i].find(class_='my-3 flex items-center') is None:
        zdalnie = 'Nie'
    else:
        zdalnie = 'Tak'

    if allOffers[i].find('div', class_='relative rounded-t-lg px-2 -ml-2 z-10 ') is None:
      miasto = "Nie podano"
    else:
      xd = allOffers[i].find('div', class_='relative rounded-t-lg px-2 -ml-2 z-10 ').find_all('span')
      if len(xd) != 0:
        xdd = ""
        for i in range(0, len(xd)):
          if i == 0:
            xdd = xd[i].text.strip()
          else:
            xdd = xdd + ", " + xd[i].text.strip()
        miasto = xdd
      else:
        miasto = 'Nie podano'
    
    zarobki = allOffers[i].find('p', class_='text-purple').text.strip()

    oferta['nazwa'] = allOffers[i].find('h3', class_='text-c28 font-medium mb-3 hidden md:block').text.strip()
    oferta['miasto'] = miasto
    oferta['zdalnie'] = zdalnie
    if(len(zarobki)) == 0:
      oferta['zarobki_min'] = 'Brak danych'
      oferta['zarobki_max'] = 'Brak danych'
    elif(len(zarobki) == 18):
        oferta['zarobki_min'] = zarobki[:5]
        oferta['zarobki_max'] = zarobki[8:14]
    elif(len(zarobki) == 19):
        oferta['zarobki_min'] = zarobki[:6]
        oferta['zarobki_max'] = zarobki[9:15]
    elif(len(zarobki) == 17):
        oferta['zarobki_min'] = zarobki[:5]
        oferta['zarobki_max'] = zarobki[8:13]


    ofertyInfo.append(oferta)
```
