```python
import json
import re
import pandas as pd
import urllib
from urllib.request import Request
from bs4 import BeautifulSoup
import matplotlib.pyplot as plot
from bs4 import BeautifulSoup
from urllib.request import Request, urlopen
```

```python
url1 = "http://www.salaryexplorer.com/salary-survey.php?loc="
url2 = "&loctype=1&job=1&jobtype=1"

pages = []
for page_number in range(1,244):
  print(f'Pobieranie strony {page_number}')
  with urllib.request.urlopen(f'{url1}{page_number}{url2}') as resp:
    processed_page = BeautifulSoup(resp.read().decode('utf-8'), "html.parser")
    pages.append(processed_page)
```

```python
print(pages[1].prettify())
```

```python
list_of_countries = []

start = "in "
end = " 2022"
for page in pages:
  if re.search('%s(.*)%s' % (start, end), page.find("title").text) is not None:
    list_of_countries.append(re.search('%s(.*)%s' % (start, end), page.find("title").text).group(1))

list_of_countries
```

```python
listt = []

for page in pages:
      listt = listt + page.find_all('div', class_='salaryblock floatedsalary shadow-light')

print(len(listt))
```

```python
kraje = {}

for i in range(0, 242):
    x = list_of_countries[i]
    if x == "Australia":
        kraje["Australia"] = "91200 AUD"
    elif x == "Bahamas":
        kraje["Bahamas"] = "46900 BSD"
    elif x == "Canada":
        kraje["Canada"] = "120000 CAD"
    elif x == "Cayman Islands":
        kraje["Cayman Islands"] = "39100 KYD"
    elif x == "France":
        kraje["France"] = "49800 EUR"
    elif x == "Gibraltar":
        kraje["Gibraltar"] = "40800 GIP"
    elif x == "Guernsey":
        kraje["Guernsey"] = "92600 GBP"
    elif x == "Ireland":
        kraje["Ireland"] = "38600 EUR"
    elif x == "New Zealand":
        kraje["New Zealand"] = "97700 NZD"
    elif x == "Norway":
        kraje["Norway"] = "615000 NOK"
    elif x == "Puerto Rico":
        kraje["Puerto Rico"] = "20200 USD"
    elif x == "Switzerland":
        kraje["Switzerland"] = "124000 CHF"
    elif x == "United Kingdom":
        kraje["United Kingdom"] = "71300 GBP"
    elif x == "United States":
        kraje["United States"] = "95100 USD"
    elif x == "United States Minor Outlying Islands":
        kraje["United States Minor Outlying Islands"] = "25800 USD"
    else:
        kraje[list_of_countries[i]] = listt[i].find('span', {'style': ' color: #666666;'}).text.replace(',', '').replace('yearly','').replace('( ', '').replace(' )','').replace(')','')

kraje
```

```python
import requests
from dateutil.parser import parse 
import csv

def get_all_exchange_rates_erapi(src):
    url = f"https://open.er-api.com/v6/latest/{src}"
    data = requests.get(url).json()
    if data["result"] == "success":
        exchange_rates = data["rates"]
    return exchange_rates

def convert_currency_erapi(src, dst, amount):
    exchange_rates = get_all_exchange_rates_erapi(src)
    return exchange_rates[dst] * amount

for i in range(0, 242):
    try:
        if str(kraje[list_of_countries[i]]).split()[1] != "USD":
            source_currency = str(kraje[list_of_countries[i]]).split()[1]
            destination_currency = "USD"
            amount = float(str(kraje[list_of_countries[i]]).split()[0])
            exchange_rate = convert_currency_erapi(source_currency, destination_currency, amount)

            kraje[list_of_countries[i]] = f"{exchange_rate} USD"
    except:
        print("Nie udao sie przewalutowa")
```
