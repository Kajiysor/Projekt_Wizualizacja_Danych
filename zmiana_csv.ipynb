```python
df = pd.read_csv("combined_data.csv")
df = df.drop(columns=["zarobki", "Unnamed: 0"])
df
```

```python
df = df.drop(df[df.zarobki_min == "Brak danych"].index)
df = df.drop(df[df.zarobki_max == "Brak danych"].index)
df = df.drop(df[df.miasto == "Nie podano"].index)
df['zarobki_min'] = df['zarobki_min'].str.replace(' ', '')
df['zarobki_max'] = df['zarobki_max'].str.replace(' ', '')
df['zarobki_max'] = df.zarobki_max.astype(float)
df
```

```python
df.zarobki_min = df.zarobki_min.astype(str)
df.zarobki_min = df.zarobki_min.astype(float)
df
```

```python
df['zarobki_srednie'] = (df.zarobki_min + df.zarobki_max) / 2.0
df
```

```python
df.to_csv('combined_data_with_average_salary.csv')
```
